# 散点图

```python
data = df1
df1.sort_values("total_price", ascending=False)
```

```python
scatter = (
    Scatter(init_opts=opts.InitOpts(theme="dark"))
    .add_xaxis(data["hourseSize"])
    .add_yaxis("房价", data["total_price"])
    .set_series_opts(
        label_opts=opts.LabelOpts(is_show=False),
        markpoint_opts=opts.MarkPointOpts(
            data=[
                opts.MarkPointItem(type_="max", name="最大值"),
            ]
        ),
    )
    .set_global_opts(
        legend_opts=opts.LegendOpts(is_show=False),
        title_opts=opts.TitleOpts(title="苏州 总价-面积"),
        xaxis_opts=opts.AxisOpts(
            name="面积",
            # 设置坐标轴为数值类型
            type_="value",
            # 不显示分割线
            splitline_opts=opts.SplitLineOpts(is_show=True),
        ),
        yaxis_opts=opts.AxisOpts(
            name="总价",
            name_location="middle",
            # 设置坐标轴为数值类型
            type_="value",
            # 默认为False表示起始为0
            is_scale=True,
            splitline_opts=opts.SplitLineOpts(is_show=True),
        ),
        visualmap_opts=opts.VisualMapOpts(
            is_show=False, type_="color", min_=200, max_=1000
        ),
    )
)

scatter.render_notebook()
```

```python
df2 = df1.groupby("area").agg("mean")[["unit_price"]].reset_index()
df2.unit_price = df2.unit_price.apply(lambda x: round(x, 1))
datas = [tuple(i) for i in df2.values.tolist()]
datas
```

```python
map_ = (
    Map(init_opts=opts.InitOpts(theme="bright"))
    .add("二手房均价", datas, "苏州", is_roam=False)
    .set_series_opts(
        label_opts=opts.LabelOpts(is_show=True),
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title="苏州二手房均价"),
        legend_opts=opts.LegendOpts(is_show=True), 
        tooltip_opts=opts.TooltipOpts(formatter="{b}:{c}元"),
        visualmap_opts=opts.VisualMapOpts(is_show=False, min_=3, max_=10),
    )
)

map_.render_notebook()
```

```python
bar = (
    Bar(init_opts=opts.InitOpts(theme="dark"))
    .add_xaxis(df2.area.tolist())
    .add_yaxis("均价", df2.unit_price.tolist())
    .set_series_opts(
        label_opts=opts.LabelOpts(is_show=False),
        markpoint_opts=opts.MarkPointOpts(
            data=[
                opts.MarkPointItem(type_="max", name="最大值"),
                opts.MarkPointItem(type_="min", name="最小值"),
            ]
        ),
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title="苏州二手房均价"),
        legend_opts=opts.LegendOpts(is_show=False),
    )
)

line = (
    Line(init_opts=opts.InitOpts(theme="dark"))
    .add_xaxis(df2.area.tolist())
    .add_yaxis(
        "营业额",
        df2.unit_price.tolist(),
        markline_opts=opts.MarkLineOpts(data=[opts.MarkLineItem(type_="average")]),
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title="", pos_top="48%"),
        legend_opts=opts.LegendOpts(is_show=False),
    )
)

grid = (
    Grid(init_opts=opts.InitOpts(theme="dark"))
    .add(bar, grid_opts=opts.GridOpts(pos_bottom="70%"))
    .add(line, grid_opts=opts.GridOpts(pos_top="50%"))
)

grid.render_notebook()
```

# 词云

```python
df1.fillna('')
```

```python
word_list = []
stop_words = ["花园", "业主", "出售"]
string = str("".join([i for i in df1["title"] if isinstance(i, str)]))

words = psg.cut(string)
for x in words:
    if len(x.word) == 1:
        pass
    elif x.flag in ("m", "x"):
        pass
    elif x.word in stop_words:
        pass
    else:
        word_list.append(x.word)
```

```python
data_pair = collections.Counter(word_list).most_common(100)


wc = (
    WordCloud()
    .add("", data_pair, word_size_range=[10, 100], shape="triangle")
    .set_global_opts(title_opts=opts.TitleOpts(title="二手房卖家诈骗词云图"))
)

wc.render_notebook()
```
